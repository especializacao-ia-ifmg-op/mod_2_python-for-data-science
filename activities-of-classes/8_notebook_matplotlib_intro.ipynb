### Representando erros

#### Errorbar

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
plt.errorbar(x, y, yerr=dy, fmt='ok', ecolor='green', elinewidth=3, capsize=5, capthick=3)
```

```python
from sklearn.gaussian_process import GaussianProcessRegressor

f = lambda x: np.sin(x)
x = np.array([[1, 3, 5, 6, 8]]).T
y = f(x)
```

```python
x
```

```python
y
```

```python
gp = GaussianProcessRegressor()
gp.fit(x, y)
```

```python
xmodel = np.linspace(0, 10, 1000)
```

```python
ymodel, error = gp.predict(xmodel[:, np.newaxis], return_std=True)
```

```python
plt.plot(x, y, 'or')
plt.plot(xmodel, ymodel, '-', color='grey')
plt.plot(xmodel, f(xmodel), '--', color='green')

plt.fill_between(xmodel, ymodel - error, ymodel + error, color='gray', alpha=.2)
```

### Customizando Legendas

```python
x = np.linspace(0,10,1000)
fig, ax = plt.subplots()

ax.plot(x,np.sin(x),'-b',label='Seno')
ax.plot(x,np.cos(x),'--r',label='Cosseno')
ax.axis('equal')

ax.legend(loc='lower center',ncol=2, fancybox=True,shadow=True,borderpad=0.3)
```
